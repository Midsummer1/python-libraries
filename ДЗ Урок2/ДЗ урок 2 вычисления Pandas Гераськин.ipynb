## Задание 1

```python
import numpy as np
import pandas as pd
```

```python
autors = pd.DataFrame({'autor_id' : [1, 2, 3],
                      'autor_name': ['Тургенев', 'Чехов', 'Островский']}, columns=['autor_id', 'autor_name'])
autors
```

```python
book = pd.DataFrame({'autor_id' : [1, 1, 1, 2, 2, 3, 3],
                    'book_title' : ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой',
                                    'Гроза', 'Таланты и поклонники'],
                    'price' : [450, 300, 350, 500, 450, 370, 290]}, columns=['autor_id', 'book_title', 'price'])
book
```

## Задание 2

```python
autor_price = pd.merge(autors, book, on = 'autor_id', how='outer')
autor_price
```

## Задание 3

```python
top5 = autor_price.sort_values(by='price', ascending=False).head()
top5
```

## Задание 4

```python
autor_price
```

```python
gb = autor_price.groupby('autor_name')
gb
```

```python
autors_stat = pd.DataFrame({'min_price' : gb['price'].min(),
                           'max_price' : gb['price'].max(),
                           'mean_price' : gb['price'].mean()})
autors_stat
```

```python
autors_stat = autor_price.groupby('autor_name').agg({'price': ['min', 'max', 'mean']}).reset_index()
autors_stat
```

## Задание 5

```python
autor_price['cover'] = ['твердая', 'мягкая', 'мягкая', 'твердая', 'твердая', 'мягкая', 'мягкая']
autor_price
```

```python
book_info = pd.pivot_table(data=autor_price,
                           values='price',
                           index='autor_name',
                           columns='cover',
                           aggfunc='sum',
                           fill_value=0,
                           dropna=False)
book_info
```

```python
book_info.to_pickle('book_info.pkl')
```
